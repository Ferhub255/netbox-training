# Use paging to get all interfaces

```python
params = {
    'device': device_names
}
# make the initial request
response = requests.get('http://localhost:8000/api/dcim/interfaces/', headers=headers, params=params)
all_interfaces = response.json()['results']

while response.json()['next'] is not None:
    # continue to make requests as long as we are given the url for the next 'page'
    response = requests.get(response.json()['next'], headers=headers)
    # add the current results to the list
    all_interfaces.extend(response.json()['results'])
    
len(all_interfaces)
```

# Choices Endpoint

```python
response = requests.get('http://localhost:8000/api/dcim/_choices/', headers=headers)
```

```python
response.json()
```

```python
interface_types = response.json()['interface:type']

sfp_plus = None
for interface_type in interface_types:
    if interface_type['label'] == 'SFP+ (10GE)':
        sfp_plus = interface_type['value']
        break
        
# This also works:
sfp_plus = next(if_type['value'] for if_type in interface_types if if_type['label'] == 'SFP+ (10GE)')
```

```python
sfp_plus
```

# Tags - get all Top of Rack switches

```python
response = requests.get('http://localhost:8000/api/dcim/devices/?role=top-of-rack', headers=headers)
tor_switches = response.json()['results']

for tor_switch in tor_switches:
    print('{}: {}'.format(tor_switch['name'], ', '.join(tor_switch['tags'])))
```

# Filter TOR swithes by tags

```python
# zone-blue
response = requests.get('http://localhost:8000/api/dcim/devices/?role=top-of-rack&tag=zone-blue', headers=headers)
tor_switches = response.json()['results']

for tor_switch in tor_switches:
    print('{}: {}'.format(tor_switch['name'], ', '.join(tor_switch['tags'])))
```

```python
# zone-blue AND zone-red
response = requests.get('http://localhost:8000/api/dcim/devices/?role=top-of-rack&tag=zone-blue&tag=zone-red', headers=headers)
tor_switches = response.json()['results']

for tor_switch in tor_switches:
    print('{}: {}'.format(tor_switch['name'], ', '.join(tor_switch['tags'])))
```
